```
# Warning: very very slow
# this cell was broken into scraping games per year, then all the data was merged after
away_assists = []
home_assists = []
away_rebounds = []
home_rebounds = []
away_turnovers = []
home_turnovers = []

for year in range(2009, 2020):
  for month in range(1, 13):
    for day in range(1, 32):
      month_ = ("{:02d}".format(month))
      day_ = ("{:02d}".format(day))
      print("Getting game_ids from " + month_ + "/" + day_ + "/" + str(year))
      url = "https://stats.nba.com/stats/scoreboardV2?DayOffset=0&LeagueID=00&gameDate={month}%2F{day}%2F{year}".format(month=str(month_),day=str(day_), year=str(year))
      header = {'User-Agent': "test"}
      try:
        data = requests.get(url, headers = header).json()
        games = data["resultSets"][0]["rowSet"]
        for i in range(0, len(games)):
          game_id = games[i][2]
          if game_id[2] == '2' or game_id[2] == '4': # 2 reps regula season, 4 reps playoffs (1 reps preseason, 3 reps all star game)
            away_assists.append(data["resultSets"][1]["rowSet"][i*2][26])
            home_assists.append(data["resultSets"][1]["rowSet"][i*2 + 1][26])
            away_rebounds.append(data["resultSets"][1]["rowSet"][i*2][27])
            home_rebounds.append(data["resultSets"][1]["rowSet"][i*2 + 1][27])
            away_turnovers.append(data["resultSets"][1]["rowSet"][i*2][28])
            home_turnovers.append(data["resultSets"][1]["rowSet"][i*2 + 1][28])
      except requests.exceptions.RequestException as e:
        continue
      except Exception as e:
        if data == "":
          print("passed", url)
        else:
          print("Jk, this day isn't valid")
        pass
print(away_assists)
print(home_assists)
print(away_rebounds)
print(home_rebounds)
print(away_turnovers)
print(home_turnovers)
```

```
new_stats = {}
new_stats['15'] = (away_assists, home_assists, away_rebounds, home_rebounds, away_turnovers, home_turnovers)
df = pd.DataFrame(new_stats)
df = df.T
df.columns = ["away assists", "home assists",  "away rebounds", "home rebounds", "away turnovers", "home turnovers"]
```

```
# calculate stats per game


away_team_asts = []
home_team_asts = []
away_team_rebs = []
home_team_rebs = []
away_team_tovs = []
home_team_tovs = []

asts_dict = {}
rebs_dict = {}
tovs_dict = {}

for i, row in df.iterrows():
  if str(i)[-4:] == "0001":
    asts_dict = {}
    rebs_dict = {}
    tovs_dict = {}

#     ASSISTS PER GAME
  if row[0] in asts_dict:
    asts_dict[row[0]][0] += df.loc[i]['away assists']
    asts_dict[row[0]][1] += 1.0
    asts_dict[row[0]][2] = asts_dict[row[0]][0] / asts_dict[row[0]][1]
  else: 
    asts_dict[row[0]] = [df.loc[i]['away assists'], 1.0, df.loc[i]['away assists']]
   
  away_team_asts.append(asts_dict[row[0]][2])

  if row[2] in asts_dict:
    asts_dict[row[2]][0] += df.loc[i]['home assists']
    asts_dict[row[2]][1] += 1.0
    asts_dict[row[2]][2] = asts_dict[row[2]][0] / asts_dict[row[2]][1]
  else: 
    asts_dict[row[2]] = [df.loc[i]['home assists'], 1.0, df.loc[i]['home assists']]
    
  home_team_asts.append(asts_dict[row[2]][2])
  
#   REBOUNDS PER GAME
  if row[0] in rebs_dict:
    rebs_dict[row[0]][0] += df.loc[i]['away rebounds']
    rebs_dict[row[0]][1] += 1.0
    rebs_dict[row[0]][2] = rebs_dict[row[0]][0] / rebs_dict[row[0]][1]
  else: 
    rebs_dict[row[0]] = [df.loc[i]['away rebounds'], 1.0, df.loc[i]['away rebounds']]
   
  away_team_rebs.append(rebs_dict[row[0]][2])

  if row[2] in rebs_dict:
    rebs_dict[row[2]][0] += df.loc[i]['home rebounds']
    rebs_dict[row[2]][1] += 1.0
    rebs_dict[row[2]][2] = rebs_dict[row[2]][0] / rebs_dict[row[2]][1]
  else: 
    rebs_dict[row[2]] = [df.loc[i]['home rebounds'], 1.0, df.loc[i]['home rebounds']]
    
  home_team_rebs.append(rebs_dict[row[2]][2])
  
#   TURNOVERS PER GAME
  if row[0] in tovs_dict:
    tovs_dict[row[0]][0] += df.loc[i]['away turnovers']
    tovs_dict[row[0]][1] += 1.0
    tovs_dict[row[0]][2] = tovs_dict[row[0]][0] / tovs_dict[row[0]][1]
  else: 
    tovs_dict[row[0]] = [df.loc[i]['away turnovers'], 1.0, df.loc[i]['away turnovers']]
   
  away_team_tovs.append(tovs_dict[row[0]][2])

  if row[2] in tovs_dict:
    tovs_dict[row[2]][0] += df.loc[i]['home turnovers']
    tovs_dict[row[2]][1] += 1.0
    tovs_dict[row[2]][2] = tovs_dict[row[2]][0] / tovs_dict[row[2]][1]
  else: 
    tovs_dict[row[2]] = [df.loc[i]['home turnovers'], 1.0, df.loc[i]['home turnovers']]
    
  home_team_tovs.append(tovs_dict[row[2]][2])
  
df['away assists per game'] = away_team_asts
df['home assists per game'] = home_team_asts
df['away rebounds per game'] = away_team_rebs
df['home rebounds per game'] = home_team_rebs
df['away turnovers per game'] = away_team_tovs
df['home turnovers per game'] = home_team_tovs
```
